```python

sns.boxplot(data=df, palette="tab20")
plt.xticks(rotation=90)
plt.xticks(rotation=90)
plt.show()
```

```python
for x in num_col:
    Q1=df[x].quantile(.25)
    Q3=df[x].quantile(.75)
    IQR=Q3-Q1
    l_limit=Q1-1.5*IQR
    u_limit=Q3+1.5*IQR
    df[x]=df[x].clip(l_limit,u_limit)
df.boxplot()
plt.show()
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
metric_k=[]
neighbours=range(3,15)
```

```python
for k in neighbours:
    knn=KNeighborsRegressor(n_neighbors=k,metric='euclidean')
    knn.fit(x_train,y_train)
    y_preds=knn.predict(x_test)
    r2_sc=r2_score(y_test,y_preds)
    metric_k.append(r2_sc)
```

```python
plt.plot(neighbours,metric_k,'o-')
plt.xlabel("neighbours")
plt.ylabel("metric_k")
plt.show()
```

```python
knn=KNeighborsRegressor(n_neighbors=11,metric="euclidean")
knn.fit(x_train,y_train)
y_predk=knn.predict(x_test)
r2_k=r2_score(y_test,y_predk)
r2_k
```
